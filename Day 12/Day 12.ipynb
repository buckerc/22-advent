```python
import numpy as np

vals = []

with open('./input.txt', 'r') as file:
    for line in file:
        vals.append(line.strip())


height=[]
for y, line in enumerate(vals):
    height_row = []
    for x,l in enumerate(line):
        if ord(l)==ord('S'):
            height_row.append(ord('a'))
            start = [y,x]
        elif ord(l)==ord('E'):
            height_row.append(ord('z'))
            end = [y,x]
        else:
            height_row.append(ord(l))
    height.append(height_row)
height = np.array(height)
current_y,current_x = start
end_y, end_x = end

max_y, max_x = height.shape
max_y -= 1
max_x -= 1
```

```python
max_steps = 1e30


def checker(position):
    current_y, current_x = position

    valid_move = []

    if current_x > 0:
        if (height[current_y, current_x-1] <= height[current_y,current_x]) or \
           ((height[current_y, current_x-1]-1) <= height[current_y,current_x]):
            valid_move.append([current_y, current_x-1])

    if current_x < max_x:
        if (height[current_y, current_x+1] <= height[current_y,current_x]) or \
           ((height[current_y, current_x+1]-1) <= height[current_y,current_x]):
            valid_move.append([current_y, current_x+1])

    if current_y > 0:
        if (height[current_y-1, current_x] <= height[current_y,current_x]) or \
           ((height[current_y-1, current_x]-1) <= height[current_y,current_x]):
            valid_move.append([current_y-1, current_x])

    if current_y < max_y:
        if (height[current_y+1, current_x] <= height[current_y,current_x]) or \
           ((height[current_y+1, current_x]-1) <= height[current_y,current_x]):
            valid_move.append([current_y+1, current_x])

    return valid_move

have_been = np.array([[current_y, current_x]])
valid_moves = checker([current_y, current_x])

counter = 0
steps = -1
walker = {}
for vm in valid_moves:
    walker[counter] = {}
    walker[counter]['steps'] = steps+1
    walker[counter]['x'] = vm[1]
    walker[counter]['y'] = vm[0]   
    walker[counter]['been'] = have_been
    counter += 1

df = pd.DataFrame(walker).T

iteration = 0
keys = list(df.index)  
while len(df) > 0:
    iteration += 1
    if iteration % 50 == 0:
        print(iteration, len(keys))
    if iteration > 400:
        break

    new_walker = {}
    counter = 0
    for row in df.iloc():
        #for key in keys:
        valid_moves = checker([row['y'], row['x']])

        for vm in valid_moves:

            if (vm == row['been'][...,:2]).all(1).any():
                continue

            if vm[0] == end_y and vm[1]== end_x:

                if max_steps >= row['steps']+1:
                    max_steps = row['steps']+1 
                    print(max_steps+1)
                    #assert 3==2
            
            counter += 1
            new_walker[counter] ={}
            new_walker[counter]['steps'] = row['steps'] + 1
            new_walker[counter]['x'] = vm[1]
            new_walker[counter]['y'] = vm[0]     
            new_walker[counter]['yx'] = str(vm[0]) + str(vm[1])
            new_walker[counter]['been'] = np.concatenate([row['been'], [[row['y'],row['x']]]])

    df = pd.DataFrame(new_walker).T
    if len(df) > 0:
        df = df.sort_values('steps')
        df = df.drop_duplicates('yx')
        df = df.reset_index(drop=True)    
        
    keys = list(df.keys())
    
max_steps+1
```
