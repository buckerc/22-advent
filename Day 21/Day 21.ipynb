```python
test = '''root: pppw + sjmn
dbpl: 5
cczh: sllz + lgvd
zczc: 2
ptdq: humn - dvpt
dvpt: 3
lfqf: 4
humn: 5
ljgn: 2
sjmn: drzm * dbpl
sllz: 4
pppw: cczh / lfqf
lgvd: ljgn * ptdq
drzm: hmdt - zczc
hmdt: 32'''

vals = []
for line in test.split('\n'):
    first = ''
    second = ''
    opp = ''
    num = 0
    monkey = line.split(': ')[0]
    rest = line.split(': ')[1]
    if len(rest.split(' ')) > 1:
        first = rest.split(' ')[0]
        opp = rest.split(' ')[1]
        second = rest.split(' ')[2]
    else:
        num = int(rest)
    vals.append({
        'monkey': monkey,
        'first': first,
        'opp': opp,
        'second': second,
        'num':num
        
    })
df = pd.DataFrame(vals)
df
```
