```python
def get_substring(search_string,idx=0):
    sub = []
    keep_going=False
    idx
    temp_sub = []
    while idx < (len(search_string)-1):
        ss = search_string[idx]
    
        if ss.isnumeric() and keep_going == False:
            num_idx = idx
            keep_going = True
        elif ss.isnumeric():
            pass
        else:
            if keep_going:
                if (num_idx - (idx-1))==0:
                    temp_sub.append(int(search_string[num_idx]))
                else:
                    temp_sub.append(int(search_string[num_idx:idx]))
                keep_going = False                
            
            if ss == ']':
                sub.append(temp_sub)
                temp_sub=[]
                break
            elif ss == '[':
                keep_going = False
                idx_sub, temp_sub_sub = get_substring(search_string[idx+1:])
                if len(temp_sub_sub)==0:
                    temp_sub.append(temp_sub_sub)
                else:
                    temp_sub.append(temp_sub_sub[0])

                #print(idx, idx_sub)
                idx += idx_sub+1
        idx += 1
        
    if keep_going:
        if (num_idx - (idx-1))==0:
            temp_sub.append(int(search_string[num_idx]))
        else:
            temp_sub.append(int(search_string[num_idx:idx]))        
        sub.append(temp_sub)
        temp_sub = []
    
    if len(temp_sub) >0 :
        sub.append(temp_sub)

    return idx, sub
            
```

```python
def compare(ll,rr):

    if type(ll)==int and type(rr)==int:
        if ll<rr:
            #print('winning')
            return 1
        elif ll>rr:
            #print('losing')
            return -1
        else:
            #print("don't know")
            return 0
    elif type(ll)==int and type(rr)!=int:    
        ll = [ll,]                
        return compare(ll,rr)        
    elif type(ll)!=int and type(rr)==int:
        rr = [rr],
        return compare(ll,rr)
    elif type(ll)!=int and type(rr)!=int:

        for l,r in zip(ll,rr):
            c = compare(l,r)
            if c == 1:
                return 1
            elif c== -1:
                return -1
        else:
            len_l = len(ll)
            len_r = len(rr)
            if len_l < len_r:
                return 1
            elif len_l > len_r:
                return -1
            else:
                return 0
                assert 3==2, "WTF2"        
    else:
        return 0
        assert 3==2, "WTF"
        
tots = 0
for counters, (left,right) in enumerate(zip(lefts,rights)):
    _, ll = get_substring(left)
    _, rr = get_substring(right)

    ll = ll[0][0]
    rr = rr[0][0]
    c = compare(ll,rr)
    if c == 1:
        tots += (counters +1)
    elif c==0:
        assert 3==2, "WTF3"
tots
```
