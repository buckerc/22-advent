```python
landscape = np.zeros((int(15000+1),7), dtype=int)

landscape[0,:]=1
winning = False
val_idx = 0
piece_idx = 0

while piece_idx < 2022:

    piece='falling'
    bottom = np.argwhere(landscape==1)[:,0].max()+4
    if piece_idx % 5 == 0:
        blocks = np.array([[bottom,2], [bottom,3], [bottom,4], [bottom,5]])
    elif piece_idx % 5 == 1:
        blocks = np.array([[bottom+2,3], [bottom+1,2], [bottom+1,3], [bottom+1,4], [bottom,3]])
    elif piece_idx % 5 == 2:
        blocks = np.array([[bottom+2,4], [bottom+1,4], [bottom,4], [bottom,3], [bottom,2]])
    elif piece_idx % 5 == 3:
        blocks = np.array([[bottom+3,2], [bottom+2,2], [bottom+1,2], [bottom,2]])
    elif piece_idx % 5 == 4:
        blocks = np.array([[bottom+1,2], [bottom+1,3], [bottom,2], [bottom,3]])  
    #assert 3==2
    while piece=='falling':                
          
        if vals[val_idx % len(vals)] == ">": #move right               
            if ((blocks[...,1] + 1) < 7).all():
                can_hoz = True
                for b in blocks:
                    if landscape[b[0],b[1]+1] != 0:
                        can_hoz = False     
                        
                if can_hoz:
                    blocks[:,1] += 1  
                
            
        elif vals[val_idx % len(vals)] == "<": #move left
            if ((blocks[...,1] - 1) >= 0).all():
                can_hoz = True
                for b in blocks:
                    if landscape[b[0],b[1]-1] != 0:
                        can_hoz = False
                if can_hoz:
                    blocks[:,1] -= 1                         
            
 
        # move vert

        can_vert = True
        for b in blocks:
            if landscape[b[0]-1,b[1]] != 0:
                can_vert = False
        if can_vert:
            blocks[:,0] -= 1
        else:
            piece = 'not falling'
            for b in blocks:
                landscape[b[0],b[1]]=1 
                
        val_idx = val_idx + 1
        
    piece_idx = piece_idx + 1
         

np.argwhere(landscape==1)[:,0].max()
```
