```python
vals = []

with open('./input.txt', 'r') as file:
    for line in file:
        vals.append(line.strip())
```

```python
start_counter = False
folders = {}
folders['check'] = [] # we have to add the current file size to paths in this list
folders['been'] = []  # have we done an `$ ls` on this path before?
path = []             # the current path

for idx, v in enumerate(vals):    
    if v.strip() == '$ cd ..':
        folders['check'].pop(-1)
        start_counter = False        
        _ = path.pop(-1)
        
        full_path = ''
        for p in path:
            full_path += f'{p}'
            
    elif '$ cd ' in v:    
        start_counter = False
        
        path.append(f"{v.split(' ')[-1]}/")
        full_path = ''
        for p in path:
            full_path += f'{p}'

        if full_path not in folders.keys():
            folders['check'].append(full_path)
            folders[full_path] = 0

    elif '$ ls' in v:
        if full_path not in folders['been']:
            folders['been'].append(full_path)
            start_counter = True            
        
    elif v[0].isnumeric() and start_counter:
        for f in folders['check']:
            folders[f] += int(v.split(' ')[0])

_ = folders.pop('check', None)
_ = folders.pop('been', None)
```

```python
# Q 1
tots = 0
for k,v in folders.items():
    if v<100_000:
        tots += v
tots
```
